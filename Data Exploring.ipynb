Importing and inspecting the data

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
clean = pd.read_csv('Cleaned_Data.csv',index_col=0)
```

```python
clean.head()
```

```python
clean.info()
```

```python
len(clean[clean['SQL'] == 1])
```

<font size="6">Let's Explore the Data !!!</font>

Top 10 Cities for Data Analyst Jobs

```python
top_cities_loc = clean['Location'].value_counts().head(10)
plt.figure(figsize=(15,5))
sns.barplot(x=top_cities_loc.index, y=top_cities_loc)
plt.title('Top 10 Cities for Data Analyst Jobs')
```

As you can see New York City has the most Data Analyst job vacancies

Top 10 States for Data Analyst Jobs

```python
top_states_loc = clean['Job_State'].value_counts().head(10)
plt.figure(figsize=(15,5))
sns.barplot(x=top_states_loc.index, y=top_states_loc)
plt.title('Top 10 States for Data Analyst Jobs')
```

As you can see California has the most Data Analyst job vacancies

Top 10 Headquarters in Cities for Data Analyst Jobs

```python
top_hq_city_loc = clean['Headquarters'].value_counts().head(10)
plt.figure(figsize=(15,5))
sns.barplot(x=top_hq_city_loc.index, y=top_hq_city_loc)
plt.title('Top 10 Headquarters in Cities for Data Analyst Jobs')
```

As you can see New York City has the most Data Analyst Job Headquarters

Top 10 Headquarters in the Same State as Job Location for Data Analyst Jobs

```python
top_hq_state_loc = clean['HQ_Same_State'].value_counts().head(11)
top_hq_state_loc.drop('na', inplace=True)
plt.figure(figsize=(15,5))
sns.barplot(x=top_hq_state_loc.index, y=top_hq_state_loc)
plt.title('Top 10 Headquarters in the Same State as Job Location for Data Analyst Jobs')
```

As you can see California has the most Data Analyst Job Headquarters

Size of Employees in Every Companies

```python
clean['Size'].value_counts().plot(kind = 'barh', color='red')
```

Most companies tend to have 10000+ employees, it could be because of the age of the company

Average Data Analyst Salary of every Companies

```python
plt.figure(figsize=(15,5))
sns.histplot(clean['Avg_Salary'])
plt.title('Average Data Analyst Salary of every Companies')
```

The Average Data Analyst Salary of every company is around 50K to 80K USD

Company Revenues based on Age of the Company

```python
plt.figure(figsize=(10,5))
sns.stripplot(x='Revenue',y='Age',data=clean)
plt.title('Company Revenues based on Age of the Company')
plt.xticks(rotation=90)
```

As you can see age isn't really the biggest factor when it comes to revenues, a lot of company's age below 50 years old still have substantial revenues

Average Data Analyst Salary based on Seniority

```python
sns.boxplot(x='Job_Seniority', y='Avg_Salary', data=clean)
plt.title('Average Data Analyst Salary based on Seniority')
```

If you're applying for Senior or Junior Data Analyst, you'd still get a decent salary above 50K USD

Job Ratings based on Sectors

```python
plt.figure(figsize=(15,5))
sns.barplot(x='Sector',y='Rating',data=clean)
plt.title('Job Ratings based on Sectors')
plt.xticks(rotation=90)
```

If you're interested in Information Technology or Travel & Tourism or Education Sector, there is a high chance that you'd work for a high rated company/institution

Job Ratings based on Industries

```python
plt.figure(figsize=(20,10))
sns.barplot(x='Industry',y='Rating',data=clean)
plt.title('Job Ratings based on Industries')
plt.xticks(rotation=90)
```

If you want to work in Architectural Engineering Services or Colleges & Universities or IT Services Industry, there is a high chance that you'd work for a high rated company/institution

Most Used Tools for Data Analyst Jobs

```python
tools = ['Python', 'Tableau','Spark', 'PowerBI', 'Excel', 'SQL']
total_tools = []
[total_tools.append(len(clean[clean[tool] == 1])) for tool in tools]

plt.figure(figsize=(15,5))
sns.barplot(x=tools, y=total_tools)
plt.title('Most Used Tools for Data Analyst Jobs')
```

Average Salary of Data Analyst Jobs based on the Comparison of two of the Most Used Tools

```python
g = sns.FacetGrid(clean, col='Excel', row='SQL', height=4)
g = g.map(sns.histplot, 'Avg_Salary')
g.fig.subplots_adjust(top=0.9)
g.fig.suptitle('Average Salary of Data Analyst Jobs based on the Comparison of two of the Most Used Tools')
```

If you have professional skills on both tools, there seems to be a lot of Data Analyst jobs with the average salary of 60K to 80K USD
